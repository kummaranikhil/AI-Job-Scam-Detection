```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv(
    "../data/processed/cleaned_jobs.csv"
)
```

```python
X_text = df["text"]
```

```python
y = df["fraudulent"]
```

```python
tfidf = TfidfVectorizer(
    max_features=5000
)
```

```python
X = tfidf.fit_transform(X_text)
```

```python
X.shape
```

```python
import joblib
```

```python
joblib.dump(
    tfidf,
    "../models/tfidf_vectorizer.pkl"
)
```

```python
df["description_length"] = (
    df["description"]
    .fillna("")
    .apply(len)
)
```

```python
df["has_company_profile"] = (
    df["company_profile"]
    .notnull()
    .astype(int)
)
```

```python
df["requirements_length"] = (
    df["requirements"]
    .fillna("")
    .apply(len)
)
```

```python
structured_features = df[
    [
        "description_length",
        "requirements_length",
        "has_company_profile"
    ]
]
```

```python
joblib.dump(
    structured_features.columns.tolist(),
    "../models/feature_columns.pkl"
)
```
